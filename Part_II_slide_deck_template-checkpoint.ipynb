## Investigation Overview

In this investigation, i'm more interested in finding out about overall distribution of Passenger's satisfaction level to determine how well airlines satisfy their passengers and also the distribution of the passenger's travel class.


## Dataset Overview

The dataset contains the airline satisfaction record for 129,880 passengers where each record represents one passenger details. The record details of a passenger covers passenger demographics, delay, travel class, flight distance, purpose, ratings for factors such as comfort, cleanliness, service.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv ('airline_cleaned.csv')
```

## Passengers Satisfaction Distribution

We can see that Passenger overall satisfaction level which is the focus on this dataset shows that a passenger can either be Satisfied/Neutral or Unsatisfied

```python
#creating color scheme
default_color = sb.color_palette()[0]
sb.countplot(data = df, x = 'Satisfaction', color = default_color)
plt.xlabel('Satisfaction')
plt.title('Passenger Satisfaction')
```

## Departure Delay vs Arrival Delay

It can be noticed that Departure delay and Arrival delay has a linear relationship

```python
# plot for Departure delay and Arrival delay 
plt.figure(figsize = [8, 6])
plt.scatter(data = df, x = 'Departure Delay', y = 'Arrival Delay', alpha = 1/5)
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.title('Scatter plot of Departure Delay vs. Arrival Delay')
plt.show()
```

## Passenger Class

It is noticed from this dataset that there are three Travel class namely Business class, Economy Class and Enomomy plus with Business class having the higest count.

```python
default_color = sb.color_palette()[0]
sb.countplot(data = df, x = 'Class', color = default_color)
plt.xlabel('Class')
plt.title('Passenger Class')
```
